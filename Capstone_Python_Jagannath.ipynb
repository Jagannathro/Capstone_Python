```python
#Task 1: Three different tables must be turned into dataframes and then saved as CSV files.
```

```python
import pandas as pd
import numpy as np

# DataFrame 1: Product DataFrame
data1 = {
    'ID': ['A001','A002','A003','A004','A005','A002','A005','A003','A001','A003','A001','A004','A004','A005'],
    'Project': ['Project 1','Project 2','Project 3','Project 4','Project 5','Project 6','Project 7','Project 8','Project 9','Project 10','Project 11','Project 12','Project 13','Project 14',],
    'Cost': [1002000,2000000,4500000,5500000,np.nan,680000,400000,350000,np.nan,300000,2000000,1000000,3000000,200000],
    'Status': ['Finished','Ongoing','Finished','Ongoing','Finished','Failed','Finished','Failed','Ongoing','Finished','Failed','Ongoing','Finished','Finished']
}
df1 = pd.DataFrame(data1)
df1.to_csv('products_data.csv', index=False)

# DataFrame 2: Employee DataFrame
data2 = {
    'ID': ['A001','A002','A003','A004','A005'],
    'Name': ['John Alter', 'Alice Luxumberg', 'Tom Sabestine', 'Nina Adgra', 'Amy Johny'],
    'Gender': ['M', 'F', 'M', 'F', 'F'],
    'City': ['Paris','London','Berlin','Newyork','Madrid'],
    'Age': [25,27,29,31,30]
}
df2 = pd.DataFrame(data2)
df2.to_csv('employee_data.csv', index=False)

# DataFrame 3: Seniority Level DataFrame
data3 = {
    'ID': ['A001','A002','A003','A004','A005'],
    'Designation_Level': ['2', '2', '3','2','3'],
    }
df3 = pd.DataFrame(data3)
df3.to_csv('seniority_data.csv', index=False)

print("CSV files have been created successfully.")
```

```python
#Task 2: Replace the missing values by running average
```

```python
# Read the CSV file
products_data = pd.read_csv('products_data.csv')

# Calculate the running average and fill NaN values
sum_cost = 0
count = 0

for index, row in products_data.iterrows():
    if not pd.isna(row['Cost']):
        sum_cost += row['Cost']
        count += 1
    else:
        if count > 0:  # To avoid division by zero
            running_avg = sum_cost / count
            products_data.at[index, 'Cost'] = running_avg

# Setting display format for better readability
pd.options.display.float_format = '{:.2f}'.format

# Save the updated DataFrame to the same CSV file
products_data.to_csv('products_data.csv', index=False)

# Print the updated DataFrame
products_data
```

```python
#Task 3: Split the name column in the Employee dataframe into two new columns “First Name”, and “Last Name” and remove the older “name” column
```

```python
# Read the CSV file
employee_data = pd.read_csv('employee_data.csv')

# Split the 'Name' column into 'First Name' and 'Last Name'
employee_data[['First Name', 'Last Name']] = employee_data['Name'].str.split(n=1, expand=True)

# Remove the old 'Name' column
employee_data.drop(columns=['Name'], inplace=True)

# Save the updated DataFrame to the same CSV file
employee_data.to_csv('employee_data.csv', index=False)

employee_data
```

```python
#Task 4: Join all three dataframes in one single dataframe. Name it as “Final”
```

```python
# Read the Products data CSV file
products_data = pd.read_csv('products_data.csv')
employee_data = pd.read_csv('employee_data.csv')
seniority_data = pd.read_csv('seniority_data.csv')

# Merge products_data, employee_data and seniority_data
Final = products_data.merge(employee_data, on='ID').merge(seniority_data, on='ID')

# Save the updated DataFrame to the CSV file
Final.to_csv('Final_data.csv', index=False)

# Display the final DataFrame
print("Final DataFrame:")
Final
```

```python
#Task 5: Added a new 'Bonus' column in the Final dataframe.
```

```python
# Read the Final data CSV file
Final = pd.read_csv('Final_data.csv')

# Add a new 'Bonus' column, giving a 5% bonus on the project cost for finished projects
Final['Bonus'] = Final.apply(lambda row: row['Cost'] * 0.05 if row['Status'] == 'Finished' else 0, axis=1)

# Save the updated DataFrame to the CSV file
Final.to_csv('Final_data.csv', index=False)

# Display the final DataFrame after adding bonus
print("Final DataFrame after adding bonus:")
Final
```

```python
#Task 6: Demote the designation level by 1, whose projects have status “fail”.
```

```python
# Read the Final data CSV file
Final = pd.read_csv('Final_data.csv')

# Demote designation level by 1 for failed projects 
Final.loc[Final['Status'] == 'Failed', 'Designation_Level'] += 1

#filter the data whose the designation level is <=4
Final = Final[Final['Designation_Level']<= 4]

# Save the updated DataFrame to the CSV file
Final.to_csv('Final_data.csv', index=False)

Final
```

```python
#Task 7: Added “Mr.” and “Mrs.” to the first name column and drop the gender column
```

```python
# Read the Final data CSV file
Final = pd.read_csv('Final_data.csv')

# Add title based on gender
Final['First Name'] = Final.apply(lambda row: 'Mr. ' + row['First Name'] if row['Gender'] == 'M' else 'Mrs. ' + row['First Name'], axis=1)

# Drop the gender column
Final.drop(columns=['Gender'], inplace=True)

# Save the updated DataFrame to the CSV file
Final.to_csv('Final_data.csv', index=False)

Final
```

```python
#Task 8: Promoted designation level by 1 for the employees whose age is more than 29 years.
```

```python
# Read the Final data CSV file
Final = pd.read_csv('Final_data.csv')

# Increase designation level by 1 for employees aged more than 29
Final.loc[Final['Age'] > 29, 'Designation_Level'] -= 1

# Save the updated DataFrame to the CSV file
Final.to_csv('Final_data.csv', index=False)

Final
```

```python
#Task 9: Calculate the total cost of all projects for each employee
```

```python
# Read the Final data CSV file
Final = pd.read_csv('Final_data.csv')

# Group by EmployeeID and FirstName and sum the project costs
total_proj_cost = Final.groupby(['ID', 'First Name']).agg({'Cost': 'sum'}).reset_index()

# Rename the 'Cost' column to 'Total cost'
total_proj_cost.rename(columns={'Cost': 'Total cost'}, inplace=True)

# Save the updated DataFrame to the CSV file
total_proj_cost.to_csv('Total_proj_cost.csv', index=False)

total_proj_cost
```

```python
#Task 10: All the employee details whose city name contains the letter “o” in it
```

```python
# Read the Final data CSV file
Final = pd.read_csv('Final_data.csv')

# Filter the DataFrame based on whether the "City" column contains the letter "o"
filtered_data = Final[Final['City'].str.contains('o')]

# Save the updated DataFrame to the CSV file
filtered_data.to_csv('Filtered_data.csv', index=False)

filtered_data
```
